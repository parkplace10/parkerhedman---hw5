```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
schools = school_data_complete['school_name'].unique()
students = school_data_complete['Student ID'].unique()
budget = school_data['budget'].sum()
avg_math_score = student_data['math_score'].mean()
avg_reading_score = student_data['reading_score'].mean()
avg_ovr_score = ((avg_math_score + avg_reading_score)/2)/100
students_passing_math = student_data[student_data['math_score']>=70].count()['math_score']
percent_passing_math = students_passing_math / len(students)
students_passing_reading = student_data[student_data['reading_score']>=70].count()['reading_score']
percent_passing_reading = students_passing_reading / len(students)

summary = {'Total Schools':[len(schools)],'Total Students':[len(students)],
          'Total Budget':['${:,.0f}'.format(budget)],'Avg Math Score':[avg_math_score],
          'Avg Reading Score':[avg_reading_score],'% Students Passing Math':['{:,.1%}'.format(percent_passing_math)],
          '% Students Passing Reading':['{:,.1%}'.format(percent_passing_reading)],"% Overall Passing":['{:,.1%}'.format(avg_ovr_score)]}
district_summary = pd.DataFrame(data=summary)
district_summary

```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#create new dataframe
school_summary_df = school_data_complete[['school_name','type','size','budget','reading_score','math_score']].copy()
#add overall scores
school_summary_df['overall_score'] = (school_summary_df['reading_score'] + school_summary_df['math_score'])/2
school_summary_df

#get school score averages & count of students passing
school_reading_avgs = pd.DataFrame(school_summary_df.groupby('school_name')['reading_score'].mean())
school_math_avgs = pd.DataFrame(school_summary_df.groupby('school_name')['math_score'].mean())
school_overall_avgs = pd.DataFrame(school_summary_df.groupby('school_name')['overall_score'].mean())

#widdle dataframe down to something that makes sense
school_scores = school_reading_avgs.merge(school_math_avgs,on='school_name',how='inner')
school_scores = school_scores.merge(school_overall_avgs,on='school_name',how='inner')
school_summary_df = school_summary_df.merge(school_scores,how='inner',on='school_name')
school_summary_df = school_summary_df.drop(columns=['reading_score_x','math_score_x','overall_score_x'])
school_summary_df = school_summary_df.drop_duplicates('school_name')
school_summary_df["Per Student Budget"] = school_summary_df['budget']/school_summary_df['size']
school_summary_df

```
